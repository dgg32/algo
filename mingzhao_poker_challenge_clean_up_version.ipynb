```
import numexpr
import copy

numexpr.evaluate('13 + 6/2 * 6').item()

operator_pair = {"+": "-", "-": "+", "*": "/", "/": "*"}
```

```
import math



##example forward [5, "-", 1, "/", 5, "*", 5]
## reverse [5, "*", 5, "/", 1, "-", 5]

def recurse(part_sum, sequence, temp_record):
    if len(sequence) == 1:
        if math.isclose(sequence[0], part_sum, abs_tol=0.00000000000001):
            temp_record.append("+")
            return [True, temp_record]
        elif math.isclose(sequence[0], -part_sum, abs_tol=0.00000000000001):
            temp_record.append("-")
            return [True, temp_record]
        else:
            temp_record.append("+")
            return [False, temp_record]
    else:
        last_digit = sequence[-1]
        temp = sequence[:-1]
        make_it = False

        successful_record = []

        for operator in ["+", "-", "*", "/"]:
            anti_operator = operator_pair[operator]

            new_target = numexpr.evaluate(str(part_sum) + anti_operator + str(last_digit)).item()
            
            new_record = copy.deepcopy(temp_record)

            new_record.append(operator)

            #print ((new_target, temp, new_record))

            [result, small_record] = recurse(new_target, temp, new_record)

            if result == True:
                make_it = True
                successful_record = small_record

                #print (new_record)
                break

        if make_it == True:
            return [True, successful_record[::-1]]
        else:
            return [False, []]





##test

#assert recurse(5, [5]) == True

#assert recurse(8, [5]) == False

#assert recurse(7, [3, 4]) == True

#assert recurse(7, [2, 1]) == False

#print (recurse(4, [2, 1]))

#print (recurse(5, [3, 7]))

#print (recurse(24, [2, 3, 2, 2], []))

#print (recurse(100, [5, 5, 1, 5], []))

print (recurse(100, [5, 1, 5, 5], []))

print (recurse(24, [1, 5, 5, 5], []))

print (recurse(7, [1, 1, 10, 3], []))
```

```
target = 24

ok_sequences = set()
all_combi = set()

for i in range(1, 14):
    print ("i", i)
    for j in range(1, 14):
        print ("j", j)
        for k in range(1, 14):
            for l in range(1, 14):
                #sequences.append([i, j, k, l])
                #print (recurse(24, [2, 3, 2, 2]))
                result = recurse(target, [i, j, k, l], [])[0]

                if result == True:
                    #print ([i, j, k, l])

                    ok_sequences.add(tuple(sorted([i, j, k, l])))

                all_combi.add(tuple(sorted([i, j, k, l])))

#print (sequences[:10])

print ("OK:", len(ok_sequences))
print ("Total draw:", len(all_combi))

print ("Percentage:", 100 * len(ok_sequences) / len(all_combi))
```
